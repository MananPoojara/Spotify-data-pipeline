# Cleaning Data for what actually needed

## Album level data cleaning 

```python
album_list = []
for row in data['items']:
    album_id = row['track']['album']['id']
    album_name = row['track']['album']['name']
    album_release_year = row['track']['album']['release_date']
    album_total_tracks = row['track']['album']['total_tracks']
    album_url = row['track']['album']['external_urls']['spotify']
    album_element = {'album id':album_id,'album name':album_name,'album release year':album_release_year,'album total tracks':album_total_tracks,'album url':album_url}
    album_list.append(album_element)
```

```python
album_list
```

## Artist level Data cleaning

```python
#single artist for one track
data['items'][2]['track']['artists']
```

```python
#multiple artist like list of artist 
data['items'][4]['track']['artists']
```

```python
artist_list = []
for row in data['items']:
    for key, value in row.items():
        if key == "track":
            for artist in value['artists']:
                artist_element = {'artist_id':artist['id'],'artist_name':artist['name'],'extrnal url':artist['external_urls']['spotify']}
                artist_list.append(artist_element)
```

```python
artist_list
```

## Song level Data Celaning

```python
data['items'][0]['track']['name']
data['items'][0]['track']['id']
data['items'][0]['track']['duration_ms']
data['items'][0]['track']['external_urls']
data['items'][0]['track']['popularity']
data['items'][0]['added_at']
```

```python
song_list = []

for row in data['items']:
    song_name = row['track']['name']
    song_id = row['track']['id']
    song_duration_time = row['track']['duration_ms']
    song_popularity = row['track']['popularity']
    song_url = row['track']['external_urls']['spotify']
    song_added = row['added_at']
    song_name = row['track']['name']
    album_id = row['track']['album']['id']
    artist_id = row['track']['album']['artists'][0]['id']
    song_element = {'song_name':song_name,'song_id':song_id,'song_duration_time':song_duration_time,'song_popularity':song_popularity,'song_url':song_url,'song_added':song_added,'song_name':song_name,'album_id':album_id,'artist_id':artist_id}
    
    song_list.append(song_element)
```
